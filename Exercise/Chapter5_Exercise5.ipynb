```python
import numpy as np
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

### Load the image data, get a sample plot.

```python
digit_data = np.loadtxt('image_digits.txt')

image_idx = np.random.random_integers(1, 1000, 49)

plt.figure(1)

for i in np.arange(1,50):
    plt.subplot(7, 7, i)
    plt.xticks(())
    plt.yticks(())
    img = digit_data[image_idx[i-1],:].reshape((28,28)).T
    plt.imshow(img)
    
```

## RGB --> Grayscale

```python
plt.figure(2)

for i in np.arange(1,50):
    plt.subplot(7, 7, i)
    plt.xticks(())
    plt.yticks(())
    img1 = rgb2gray(digit_data[image_idx[i-1],:].reshape((28,28)).T)
    plt.imshow(img1)
    
```
